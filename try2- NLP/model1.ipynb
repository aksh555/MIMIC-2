```python
input1 = Input(shape=(maxlen,))
print(input1.shape)
 
x = Embedding(max_words, 20, input_length=maxlen)(input1)

x = squash_function()(x)

x = Conv1D(96, kernel_size=3, padding="valid", kernel_initializer="glorot_uniform")(x)

x = squash_function()(x)
print(x.shape)

#m = conv_2_layer()(x)
conv1 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv2 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv3 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat = concatenate([conv2,conv1])
cat2 = concatenate([cat,conv3])

conv_1 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_2 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_3 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat_1 = concatenate([conv_2,conv_1])
cat_2 = concatenate([cat_1,conv_3])

m = concatenate([cat2,cat_2])

#n = conv_2_layer()(x)
conv1 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv2 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv3 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat = concatenate([conv2,conv1])
cat2 = concatenate([cat,conv3])

conv_1 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_2 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_3 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat_1 = concatenate([conv_2,conv_1])
cat_2 = concatenate([cat_1,conv_3])

n = concatenate([cat2,cat_2])

#p = conv_2_layer()(x)
conv1 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv2 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv3 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat = concatenate([conv2,conv1])
cat2 = concatenate([cat,conv3])

conv_1 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_2 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_3 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat_1 = concatenate([conv_2,conv_1])
cat_2 = concatenate([cat_1,conv_3])

p = concatenate([cat2,cat_2])

#q = conv_2_layer()(x)
conv1 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv2 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv3 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat = concatenate([conv2,conv1])
cat2 = concatenate([cat,conv3])

conv_1 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_2 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_3 = Conv1D(32, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat_1 = concatenate([conv_2,conv_1])
cat_2 = concatenate([cat_1,conv_3])

q = concatenate([cat2,cat_2])

l = concatenate([m,n])
o = concatenate([p,q])
x = concatenate([l,o])

x = squash_function()(x)

#a = conv_3_layer()(x)
conv1 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv2 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv3 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat = concatenate([conv2,conv1])

cat2 = concatenate([cat,conv3])

conv_1 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_2 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_3 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat_1 = concatenate([conv_2,conv_1])

cat_2 = concatenate([cat_1,conv_3])

conv4 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv5 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv6 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat3 = concatenate([conv4,conv5])

cat4 = concatenate([cat3,conv6])

l = concatenate([cat2,cat_2])
a = concatenate([l,cat4])



#b = conv_3_layer()(x)
conv1 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv2 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv3 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat = concatenate([conv2,conv1])

cat2 = concatenate([cat,conv3])

conv_1 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_2 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv_3 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat_1 = concatenate([conv_2,conv_1])

cat_2 = concatenate([cat_1,conv_3])

conv4 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv5 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
conv6 = Conv1D(16, kernel_size=2, padding="valid", kernel_initializer="glorot_uniform")(x)
cat3 = concatenate([conv4,conv5])

cat4 = concatenate([cat3,conv6])

l = concatenate([cat2,cat_2])
b = concatenate([l,cat4])

x = concatenate([a,b])

attn = AttentionWeightedAverage()(x)
max_pool = GlobalMaxPooling1D()(x)
avg_pool = GlobalAveragePooling1D()(x)
z = concatenate([max_pool,avg_pool])
x = concatenate([attn,z])

x = squash_function()(x)

preds = Dense(num_classes, activation="sigmoid")(x)
 
model = keras.Model(input1, preds)
 
model.summary()
 

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])


callbacks = [
    ReduceLROnPlateau(), 
    EarlyStopping(patience=4)
]

history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_split=0.1,
                    callbacks=callbacks)
```
